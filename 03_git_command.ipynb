## Terminal and Git Command
1. pwd: Tell us current path file, that's on.
2. ls: it shows list of folder/file.
3. cd ..: change of directory.
4. cd [name of directory]: to enter in that directory but do cd .. first to get out of the current directory
- -----------------------------
5. git init: initialize a repository
6. git status: where things are in process.
7. git add .: It add everything from that folder to git.
8. git commit -m: commit the folder along with a (-m) message.
- -------------------------------
9. git log: shows previous commit that we made. It also provides *hash code* which we can use to revert back to previous commit.
10. git checkout: In a detached head state just looking where we were at our previous commit or snapshot
11. git checkout master: It will bring you back to your most recent update/change you made to your file. 
- -------------------------------

- *You can create separate branches parallel to your master branch. You and someone else can work on their separate branch and merge it later to the master branch*.
12. git checkout -b separate_branch(name of the separate branch, can be named anything) plus the hash code from where you want to create the separate branch, we'll be using it from second commit.
- EXCERCISE: Create one more branch and name it, whatever you like.
- You can switch to brances which you create with git checkout -b {Your branch name}.
