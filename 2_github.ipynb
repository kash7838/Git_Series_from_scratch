## Github
- It is a hosting website for git. 
- It is a service.
- It is free to use, post open-source project or private.
- It is a cloud based platform.
- You can connect it to git and store things in the cloud, so, you can acces it from anywhere in the world.
- Github utilizes git for their version control. Which is how they track the changes and maintain different versions of code, etc.
- If you're on local git(your device) you can't access it from other computer but with the help of github you can, you can access it remotely from other computer.
- Github makes it easy for a team to work together.
- You can do pull code, merging, code reviews.
- It is a GUI.(Graphical User Interface)
- It is hosted on web.
- It is focused on centralized source code hosting.
- It is a hosting service for git repositories.
