## Git
- It is a distributed version control system that tracks changes in source code during software development.
- Allows multiple people to work in a project without overwriting each other's changes.
- It provides tools for branching, merging and tracking progress across phase of the project.
- It is a command-line tool.
- It is a software.
- It focuses on version control and code sharing.
- It is to manage source code history.

### Why to use?
- One the most populare version control software in the world.
- Open source(free to use)
